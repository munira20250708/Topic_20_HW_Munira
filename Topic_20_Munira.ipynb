# MODELING

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, metrics
```

```python
# Select Modelling Technique
```

Model machine learning yang digunakan adalah:
1. Logistic Regression
2. Decision Tree
3. Random Forest
4. Hyperparameter Tuning for random forest

```python
# Generate Test Design
```

```python
data1.head()
```

```python
## Check the proportion of 1 and 0 in Exited (Churn or no) label
data1['Exited'].value_counts()
```

```python
import seaborn as sns
sns.countplot(data = data1, x='Exited')
```

```python
## Headling Imbalanced Data

from sklearn.utils import resample

#create teo different dataframe of majority and minority class 
data1_majority = data1[(data1['Exited']== 0)] 
data1_minority = data1[(data1['Exited']== 1)] 

# upsample minority class
data1_minority_upsampled = resample(data1_minority,
                          replace=True,      # sample with replacement
                          n_samples=7963,     # to match majority class
                          random_state=5)   # reproducible results

# combine majority and upsampled minority
data1_upsampled = pd.concat([data1_minority_upsampled, data1_majority])
```

```python
## check new class counts
data1_upsampled['Exited'].value_counts()
```

```python
## One Hot Encoding
```

```python
data1 = pd.get_dummies(data1)
data1.head()
```

Hal ini dapat dilihat pada df.head() dengan satu metode pengkodean panas dapat mewakili atau membuat data tipe kategori sebagai vektor biner dengan nilai bilangan bulat, 0 dan 1, di mana semua elemen akan bernilai 0 kecuali untuk satu elemen yang bernilai 1, yaitu elemen yang memiliki nilai kategori tersebut.

```python
# Build Model
```

```python
data1.columns
```

```python
## defining feature matrix(x) and response vector(y)
x = data1.loc[:, data1.columns != 'Exited']
y = data1['Exited']
```

```python
## splitting X and y into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,
                                                    random_state=5)
```

```python
## Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr = lr.fit(x_train, y_train)
```

```python
## Decision Tree
```

```python
from sklearn import tree
import pydotplus
import matplotlib.pyplot as plt
import matplotlib.image as pltimg

dtree = tree.DecisionTreeClassifier()
dtree = dtree.fit(x_train, y_train)
```

```python
text_representation = tree.export_text(dtree)
print(text_representation)
```

```python
## Random Forest
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier_rf = RandomForestClassifier(random_state=5, n_jobs=-1, max_depth=5,
                                       n_estimators=100, oob_score=True)

classifier_rf.fit(x_train, y_train)
```

Kita dapat melihat bahwa  model random forest  terletak pada max_depth=5, n_jobs=1, oob_score=True, an random_state=5

```python
## Hyperparameter Tuning for random forest
```

```python
rf = RandomForestClassifier(random_state=5, n_jobs=-1)
```

```python
#Create one variable (params) to deposit whatever we will try to do with the model
params = {
    'max_depth': [2,3,5,10,20],
    'min_samples_leaf': [5,10,20,50,100,200],
    'n_estimators': [10,25,30,50,100,200]
}
```

```python
from sklearn.model_selection import GridSearchCV

# Instantiate the grid search model
grid_search = GridSearchCV(estimator=rf,
                           param_grid=params,
                           cv = 5,
                           n_jobs=-1, verbose=1, scoring="accuracy")

grid_search.fit(x_train, y_train)
```

Dari model hyperparemeter tuning for random forest, kita dapat mengetahui bahwa ada Fitting 5 folds untuk masing-masing dari 180 kandidat, dengan total 900 fit

```python
# To see best_score
grid_search.best_score_
```

```python
rf_best = grid_search.best_estimator_
rf_best.fit(x_train, y_train)
```

Kita dapat melihat bahwa best_estimator atau akurasi dalam model Hyperparameter Tuning for random forest terletak pada max_depth=10 dan min_samples_leaf=5

# EVALUATION

```python
#For X_test predictions
# Logistic Regression
y_lr = lr.predict(x_test)

# Decision Tree
y_dtree = dtree.predict(x_test)

# Random Forest Awal
y_rf_before = classifier_rf.predict(x_test)

# Random Forest dengan Hyperparameter Tuning
y_rf_after = rf_best.predict(x_test)
```

```python
## confusion_matrix
from sklearn.metrics import confusion_matrix
print("Logistic Regression : \n", confusion_matrix(y_test, y_lr))
print("Decision Tree : \n", confusion_matrix(y_test, y_dtree))
print("Random Forest  : \n", confusion_matrix(y_test, y_rf_before))
print("Random Forest dengan Hyperparameter Tuning: \n", confusion_matrix(y_test, y_rf_after))
```

Evaluasi hasil dengan confusion_matrix menunjukkan perbandingan antara BENAR POSITIF ('customer churn') dengan POSITIF PALSU (customer tidak churn) dan antara BENAR NEGATIF (customer tidak churn) dengan NEGATIF PALSU ('customer churn')

```python
## classification report
```

```python
from sklearn.metrics import classification_report
print("Logistic Regression : \n\n", classification_report(y_test, y_lr))
print("Decision Tree : \n\n", classification_report(y_test, y_dtree))
print("Random Forest : \n\n", classification_report(y_test, y_rf_before))
print("Random Forest dengan Hyperparameter Tuning: \n\n", classification_report(y_test, y_rf_after))
```

Evaluasi hasil dengan classification_report menunjukkan ringkasan beberapa metrik sehingga kita dapat melihat nilai presisi, recall, f1-score, dan akurasi.

```python
## AUC
```

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_test, y_lr, pos_label=1) # pos_label: positive label
print("Logistic Regression :", auc(fpr, tpr))
fpr, tpr, thresholds = roc_curve(y_test, y_dtree, pos_label=1) # pos_label: positive label
print("Decision Tree :", auc(fpr, tpr))
fpr, tpr, thresholds = roc_curve(y_test, y_rf_before, pos_label=1) # pos_label: positive label
print("Random Forest :", auc(fpr, tpr))
fpr, tpr, thresholds = roc_curve(y_test, y_rf_after, pos_label=1) # pos_label: positive label
print("Random Forest dengan Hyperparameter Tuning:", auc(fpr, tpr))
```

Evaluasi hasil dengan AUC menjelaskan bahwa semakin tinggi auc, semakin baik modelnya dalam membedakan antara customer churn dan tidak.

Model yang lebih baik untuk memprediksi customer churn adalah adalah 'Random Forest dengan Hyperparameter Tuning'

Setelah dilakukan evaluasi terhadap beberapa model dengan 3 metrik (confusion matrix, classification report, dan AUC), terlihat bahwa hasil prediksi dengan model 'Random Forest dengan Hyperparameter Tuning' memiliki tingkat akurasi yang paling tinggi. Dalam mengevaluasi metrik menggunakan classification_report terlihat beberapa metrik gabungan, untuk melihat seberapa akurat model kita dalam memprediksi true positive dan true negative customer churn maka metrik yang cocok adalah metrik 'recall' karena 'recall' metrik menunjukkan tingkat kebenaran model yang positif, akurasi tertinggi dari metrik recall adalah dalam prediksi model 'Random Forest dengan Hyperparameter Tuning' sebesar 85%.
